## Research Context

### Problem Background
Tuberculosis (TB) is an airborne bacterial infection caused by *Mycobacterium tuberculosis*.
Despite being preventable and curable, TB remains one of the leading causes of death from a single infectious agent worldwide.

**Nigeria's burden:**
- Nigeria ranks among the top-five highest TB-burden countries globally (WHO Global TB Report)
- Sputum microscopy — the standard diagnostic — requires laboratory infrastructure often unavailable in rural areas
- Chest X-ray (CXR) screening is faster, scalable, and amenable to AI-assisted reading

### Why AI-Assisted CXR?
Deep learning models trained on large CXR datasets can detect TB with sensitivity and specificity comparable to experienced radiologists. However:
- Training requires large datasets
- Patient data cannot be freely shared across hospitals (privacy regulations, ethical constraints)
- Models trained at one hospital may not generalise to another

### Why Federated Learning?
Federated Learning (FL) allows multiple hospitals to collaboratively train a shared model **without sharing raw patient images**:
- Each hospital trains locally on its own data
- Only model weight updates (not images) are sent to a central aggregator
- The aggregator combines updates → improved global model
- Repeat for many rounds

### Why Differential Privacy?
Even sharing weight updates can leak information about individual patients (membership inference attacks, gradient inversion). Differential Privacy (DP) adds calibrated noise to gradients, providing a formal mathematical guarantee that no individual patient can be identified from the shared updates.

### Nigerian Teaching Hospitals Context
Nigeria has ~22 federal teaching hospitals. A federated TB screening system could:
- Leverage data from all sites without centralising sensitive patient records
- Produce a model that generalises across Nigeria's heterogeneous patient populations
- Enable hospitals with limited AI expertise to benefit from collective learning

## Research Question

```python
research_question = """
MAIN RESEARCH QUESTION
======================
Can a differentially-private federated learning system trained across simulated
Nigerian teaching hospital sites achieve TB detection performance statistically
comparable to a centralised training baseline, while preserving patient privacy
and supporting cross-site generalisation?

HYPOTHESIS
==========
A federated ResNet-18 model trained with FedAvg and (ε=8, δ=1e-5)-differential
privacy will achieve:
  (1) AUC-ROC ≥ 0.85 on the held-out test set
  (2) Sensitivity ≥ 0.80 at the Youden-optimal threshold
  (3) AUC-ROC within 0.05 of the centralised baseline (non-inferiority)

SECONDARY QUESTIONS
===================
SQ1. How does model performance degrade as privacy budget ε decreases?
     (Privacy–utility trade-off analysis)

SQ2. How does data heterogeneity across sites (Dirichlet α) affect convergence?

SQ3. How many FL communication rounds are needed for convergence?

SQ4. Do individual hospital sites generalise to each other's held-out data?

SQ5. What types of errors (false negatives = missed TB) does the model make,
     and do error rates differ across simulated sites?

WHAT COUNTS AS SUCCESS
======================
- Federated + DP model: AUC-ROC 95% CI lower bound > 0.80
- Non-inferiority vs centralised: ΔAUC 95% CI contains values < 0.05
- Convergence within 50 FL rounds

WHAT COUNTS AS FAILURE
======================
- Federated AUC-ROC < 0.75 (worse than clinical utility threshold)
- Convergence requiring > 100 rounds (impractical for deployment)
- DP model AUC-ROC < 0.70 at ε=8 (DP cost too high)

WHY THESE METHODS FIT THE TASK
===============================
- ResNet-18: well-validated CXR backbone; efficient on limited compute
- FedAvg: simplest FL algorithm; strong empirical baseline for benchmarking
- Opacus DP-SGD: mature, audited PyTorch DP library
- Flower (flwr): framework-agnostic FL simulation; easy to scale to real deployment
"""

print(research_question)
```
